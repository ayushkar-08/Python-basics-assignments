Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
i=1
while i<=20:
  print(i)
  i+=1
```

Q3. Print multiplication table of 5.

Approach:
- Initialize counter
- Multiply inside loop

```python
#write answer here
i=1
while i<=10:
  result=5*i
  print(f"5*{i}={result}")
  i+=1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
i=1
result=0
n=int(input("enter n:"))
while i<=n:
  result=result+i
  print(result)
  i+=1
```
